```python
import numpy as np
import pandas as pd
import ast
```

```python
movies = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/tmdb_5000_movies.csv')
credits = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies=movies.merge(credits,on='title')
movies.head(1)
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head()
```

```python
movies.isnull().sum()
movies.dropna(inplace=True)
movies.duplicated().sum()
```

```python
def convert(text):
  L=[]
  for i in ast.literal_eval(text):
    L.append(i['name'])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def cast(text):
  L = []
  counter = 0
  for i in ast.literal_eval(text):
    if counter < 3:
      L.append(i['name'])
    counter+=1
  return L
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
movies.head()
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies= movies[['movie_id','title','tags']]
```

```python
movies.head()
```

```python
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))
```

```python
movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(movies['tags']).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
movies[movies['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = movies[movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:10]:
        print(movies.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```
